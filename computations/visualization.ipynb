```python
def hmax_immobile(a = 100, eta = 1, tau = 10e-15, R = 200e-9, wlen = 1.064e-6):
    c = 3e8
    e = 1.6e-19
    m_e = 9.1e-31
    omega = c / wlen
    crutch = 9.72 / 1.8749
    renorm = m_e * c * omega / e * 1e-11

    hrel = eta * a / (1 + a ** 2) ** (1 / 2) * c * tau / (16 * R)

    return hrel * renorm
```

```python
# get impulses

def denoise(arr: np.ndarray, cutoff: float, order: int = 4):

    import scipy.signal as signal

    B, A = signal.butter(order, cutoff, btype='low', analog=False)
    smooth_data = signal.filtfilt(B, A, arr)

    return smooth_data

# .............................. #

efield = data['Electric_Field_Ey'].copy()

# plt.imshow(efield ** 2, aspect='auto', origin='lower')
```

```python
ts = tmax * 1e-15 / efield.shape[0]
fs = 1 / ts

apply_denoise = True

tline = np.arange(0, efield.shape[0] * ts, ts)

# quite good signal, so can apply always
slice1 = denoise(slice1, cutoff = 0.15)

if apply_denoise:
    slice2 = denoise(slice2, cutoff = 0.15)

n = len(slice1)
k = np.arange(n)
T = n / fs
frq = k / T
frq = frq[range(n // 2)]

slice1fft = np.fft.fft(slice1) / n
slice1fft = slice1fft[range(n // 2)]
slice2fft = np.fft.fft(slice2) / n
slice2fft = slice2fft[range(n // 2)]
```

```python
fig, ((ax00, ax10), (ax01, ax11)) = plt.subplots(2, 2, figsize=(16, 12))

ax00.plot(tline[:n] * 1e15, slice1, linewidth=1, label='Incident pulse', color='purple')
ax00.set_xlabel(r'$t$, $\rm{fs}$', labelpad=15)
ax00.legend()


ax10.plot(tline[:n] * 1e15, slice2, linewidth=1, label='Reflected pulse')
ax10.set_xlabel(r'$t$, $\rm{fs}$', labelpad=15)
ax10.legend()

ax01.plot(frq / (3e8 / 8e-7), abs(slice1fft) / max(abs(slice1fft)), linewidth=1, color='black', label='Incident spectrum')
ax11.plot(frq / (3e8 / 8e-7), abs(slice2fft) / max(abs(slice1fft)), linewidth=1, color='black', label='Reflected spectrum')
ax01.set_xlim(0, 20)
ax11.set_xlim(0, 20)
#ax01.set_ylim(0, 1.2)
#ax11.set_ylim(0, 1)

ax01.set_yscale('log')
ax11.set_yscale('log')
ax01.legend()
ax11.legend()

ax00.set_ylabel(r'$E$, $\rm{V}\:/\:{m}$', labelpad=15)
ax10.set_ylabel(r'$E$, $\rm{V}\:/\:{m}$', labelpad=15)

ax01.set_xlabel(r'$\omega\:/\:\omega_0$', labelpad=15)
ax11.set_xlabel(r'$\omega\:/\:\omega_0$', labelpad=15)
ax01.xaxis.set_major_locator(ticker.LinearLocator(11))
ax11.xaxis.set_major_locator(ticker.LinearLocator(11))
# ax01.yaxis.set_major_locator(ticker.LinearLocator(7))
# ax11.yaxis.set_major_locator(ticker.LinearLocator(6))

plt.tight_layout()

# fig.savefig('../tex/img/pulse_spectrum.png', dpi=300, bbox_inches='tight')
```

```python
def denoise(arr: np.ndarray):

    import scipy.signal as signal

    N  = 4    # Filter order
    Wn = 0.1 # Cutoff frequency
    B, A = signal.butter(N, Wn, output='ba')
    smooth_data = signal.filtfilt(B,A, arr)

    return smooth_data
```

```python
ts = 120e-15 / ey.shape[0]
fs = 1 / ts

apply_denoise = False

tline = np.arange(0, ey.shape[0] * ts, ts)

slice1 = ey[90:1090, 200] # incident
slice2 = ey[1700:2700, 200] # reflected

# quite good signal, so can apply always
slice1 = denoise(slice1)

if apply_denoise:
    slice2 = denoise(slice2)

n = len(slice1)
k = np.arange(n)
T = n / fs
frq = k / T
frq = frq[range(n // 2)]

slice1fft = np.fft.fft(slice1) / n
slice1fft = slice1fft[range(n // 2)]
slice2fft = np.fft.fft(slice2) / n
slice2fft = slice2fft[range(n // 2)]
```
