```python
import numpy as np
import plotly.graph_objects as go

data = np.load('data/nine_history_10.npz')
ni = data['ni']
# data = np.load('data/output_ne.npz')
ne = data['ne']
```

```python
cube_slice = slice(200, 300, 1)

dataplot = ni[cube_slice, cube_slice, cube_slice]

x = np.linspace(-1500, 1500, ni.shape[0], endpoint=True)
Y, X, Z = np.meshgrid(x, x, x)
X = X[cube_slice, cube_slice, cube_slice]
Y = Y[cube_slice, cube_slice, cube_slice]
Z = Z[cube_slice, cube_slice, cube_slice]

camera = dict(
    eye=dict(x=0.1, y=2, z=0.1)
)

fig = go.Figure(data=go.Volume(
        x=X.flatten(), y=Y.flatten(), z=Z.flatten(),
        value=dataplot.flatten(),
        isomin=1e22,
        opacity=0.2,
        surface_count=25,
        colorscale='Hot',
        colorbar={"title": "n_i"}
    ))
fig.update_layout(
    template='simple_white',
    width=800,
    height=800,
    margin=dict(l=30, r=30, t=50, b=50),
    scene_camera=camera,
    font=dict(
        size=13
    )
    # scene_xaxis_showticklabels=False,
    # scene_yaxis_showticklabels=False,
    # scene_zaxis_showticklabels=False
)
fig.write_image("../tex/img/new/ni_sdf16_r200_2.png")
```
