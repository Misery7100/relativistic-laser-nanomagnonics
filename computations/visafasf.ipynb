```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker
import seaborn as sns

#! make plotting utils as submodule
from texplot import configure_mpl

configure_mpl()
```

```python
from PIL import Image, ImageOps

abc = Image.open('data/y_py_R200nm_I1022.jpg')
abc = ImageOps.grayscale(abc)
arr = np.asarray(abc)

fig, ax = plt.subplots(figsize=(10, 8))
ax.imshow(arr / 255, cmap='gray', alpha=1, extent=[-600, 600, -0.13, 0.13], aspect='auto')

ax.set_xlabel(r'$y$, $\rm{nm}$', labelpad=15)
ax.set_ylabel(r'$v_y\:/\:c$', labelpad=15)

fig.savefig(f'../tex/img/new/y_py_R200.png', dpi=300, bbox_inches='tight')
```
